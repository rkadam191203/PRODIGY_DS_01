```python
population_df = df[
    ['Country Name', '2022']
]
```

```python
population_df = population_df.dropna()
```

```python
population_df.isnull().sum()
```

```python
population_df['2022'].describe()
```

```python
plt.figure(figsize=(10,6))

plt.hist(
    population_df['2022'],
    bins=20,
    edgecolor='black'
)

plt.title('Population Distribution (2022)')
plt.xlabel('Population')
plt.ylabel('Frequency')

plt.show()
```

```python
plt.savefig(
    "histogram.png",
    dpi=300,
    bbox_inches='tight'
)
```

```python
top10 = population_df.sort_values(
    by='2022',
    ascending=False
).head(10)


    
```

```python
top10
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    data=top10,
    x='Country Name',
    y='2022'
)

plt.xticks(rotation=45)

plt.title(
    'Top 10 Most Populated Countries (2022)'
)

plt.xlabel("Country")

plt.ylabel("Population")

plt.show()
```

```python
plt.savefig(
    "top10_population.png",
    dpi=300,
    bbox_inches='tight'
)
```
